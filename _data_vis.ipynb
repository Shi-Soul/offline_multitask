```python
from util import *
from util import *
from sklearn.manifold import TSNE
# PCA
from sklearn.decomposition import PCA
# UMAP
from sklearn.manifold import MDS
dataset = make_dataset(0,1)
# merge_data = merge_dataset(*dataset.values())
merge_data = merge_dataset(dataset['walk_m'])
obs = merge_data['obs']
# obs = np.concatenate( [merge_data['obs'],merge_data['act']], axis=1)
# write a code to visualize the data, use TSNE, PCA or any other method to reduce the dimensionality of the data
# and then plot the data using scatter plot

import matplotlib.pyplot as plt

N_tot = len(obs)
N=2000
obs = obs[::N_tot//N]
# Perform dimensionality reduction using TSNE
tsne = TSNE(n_components=2)
# tsne = PCA(n_components=2)
obs_tsne = tsne.fit_transform(obs)

# Plot the data using a scatter plot
plt.scatter(obs_tsne[:, 0], obs_tsne[:, 1],c=merge_data['rew'][::N_tot//N])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Data Visualization')
# colorbar
plt.colorbar()
plt.show()
```
